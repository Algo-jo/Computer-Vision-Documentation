```python
# LIST
list1 = [1, 2, 3, 4, 5]
print(list1)
print(list1[0:4:2])

for i in list1:
    print(i, end=" ")
```

```python
# SET
set1 = set([1, 2, 3, 4, 5])
print(set1)
set1.add(10)
print(set1)
set1.pop()
print(set1)
```

```python
# Representasi JSON di Python
# Structure data menggunakan Key dan Value - Dictionary
dict1 = {
    1: "one",
    True: [1, 2, 3],
    "two": 2
}
print(dict1)
print(dict1["two"])

if dict1.get(1) is not None:
    print(dict1[1])
```

```python
# Collection looping
for i in ["bowo", "mulyo", "solo"]:
    print(i, end=" ")
print()
# Zip -> menggabungkan 2 collection yang berbeda. Jika ukuran berbeda, ikuti size collection terkecil
nama = ["budi", "ono", "siregar"]
umur = [23, 12, 190, 2000]

print(list(zip(nama, umur)))

for (nama, umur) in zip(nama, umur):
    print(f"Nama {nama}, Umur {umur}")
    
for (index, nama) in enumerate(nama):
    print(f"Index {index}, namanya {nama}")
```

```python
# DICTIONARY
dict2 = {
    "Roti" : "bread",
    "Apel": "Apple"
}
for i in dict2:
    print(i)
print(list(dict2.items()))

for (key, value) in dict2.items():
    print(f"{key}, {value}")
```
